```python
# TicTacToe_vs_ai
# This is an implementation of a program that a human play a game of Tic Tac Toe against a computer with 2 levels

# Inspiration from https://geekflare.com and https://www.thecrazyprogrammer.com
# There is a computer player in that will either make random moves (easy) or strategic moves to win (hard)

# Algorithm:

# 1. Create a board by initialising a 2d array with each elm initialised as '-'

# 2. Write a function to check whether the board is filled or not
  # iterate over the board and return False if any elm in 2d list is '-' else True

# 3. Write a function to check whether a player has won or not
  # check all 8 possible winning arrangements: all 3 rows, all 3 cols and both 2 diagonals

# 4. Write a function to show the board becuse it will be displayed throught game

# 5. Write a function to implement random computer move
  # find random number between 1 and 9 inclusive => generate loc
  # while loc is not in empty space list regenerate loc
  # once found empty loc then map to board idx and put sign there

# 6. Write a function to implement ai computer move
  # check if next move is winning for computer 
    # check if putting sign in any empty spaces makes has_player_won function True for computer sign
    # reset each time
    # if True then return that space for computer to play
  # check if next move is winning for opponent
    # check if putting sign in any empty spaces makes has_player_won function True for opponent sign 
    # reset each time
    # if True then return that space for computer to play
  # check if any of empty spaces is in corners set
  # check if any of empty spaces is in centre set
  # check if any of empty spaces is in sides set

# 7. Write function to start the game
  # Select the first turn of the player randomly
  # Write infinite loop that breaks when game over (win or draw)
    # Show board to user to select spot of next move
    # Ask the user to enter square number
    # update spot to respective player sign
    # check whether currnet player has won game or not
    # if player has won game: print winning message (output of the game) and break infinite loop
    # check whether board is filled or not (this line will run if curr player did not win game)
    # if board is filled: print draw message (output of the game) and break the infite loop
  # show the user the final view of the board
```

```python
import random
class TicTacToe:
  

  def __init__(self):
    self.board = []
    self.empty_spaces = list(range(1,10))
    self.loc_sets = [{1, 3, 7, 9}, {5}, {2, 4, 6, 8}] # idx0 set => corners , idx1 set is centre, idx2 set is sides
    self.locations = {1:(0,0), 2:(0,1), 3:(0,2), 4:(1,0), 5:(1,1), 6:(1,2), 7:(2,0), 8:(2,1), 9:(2,2)}  # locs maps to coordinates of board
  
  def create_board(self):
    self.board = [['-' for i in range(3)] for j in range(3)]
    

  def is_board_full(self):
    for row in range(len(self.board)):
      for col in range(len(self.board[0])):
        if self.board[row][col] == '-':
          return False
    return True

  def has_player_won(self, player):
    rows = len(self.board)
    cols = len(self.board[0])

    
    # all rows: 
      #for row in range(len(board)) if all elms in each row == 'x' or all == 'y'
      
    win = None

    for row in range(rows):
      win = True
      for col in range(cols):
        if self.board[row][col] != player:
          win = False
          break
      if win:
        return win
    
    #  OR
    # all cols
      #for each row if col[0] is all x or all y
      #for each row if col[1] is all x or all y
      #for each row if col[2] is all x or all y
    
    for col in range(cols):
      win = True
      for row in range(rows):
        if self.board[row][col] != player:
          win = False
          break
      if win:
        return win

    # OR
    # each diagonal
      # if 0,0 1,1 2,2 are all x or all y
      # if 0,2 1,1 2,0 are all x or all y
    win = True
    for row in range(rows):
      col = row
      if self.board[row][col] != player:
        win = False
        break
    if win:
      return win

    win = True
    for row in range(rows):
      col = cols-1-row
      if self.board[row][col] != player:
        win = False
        break
    if win:
      return win
    return False
    
    for row in self.board:
      for item in row:
        if item == '-':
          return False
    return True


  def get_random_first_player(self):
    return random.randint(0,1)
  
  # allows current player to add sign to board
  def add_sign(self, loc, player):
    
    self.empty_spaces.remove(loc)
    (row, col) = self.locations[loc]
    self.board[row][col] = player
  
  # gets next player's turn
  def next_turn(self, player):
    return 'X' if player == '0' else '0'
 
  # displays board for user so they can make their play
  def show_board(self):
    rows = len(self.board)
    cols = len(self.board[0])
    
    for row in range(rows):
      for col in range(cols):
        print(self.board[row][col], end =" ")
      print()

  def computer_random_move(self): # EASY LEVEL
    random_loc = random.randint(1,9)

  def computer_ai_move(self, player): # HARD LEVEL
    # check if next move is winning for computer 
    # check if putting sign in any empty spaces makes has_player_won function True for computer sign
    for space in self.empty_spaces:
      (row, col) = self.locations[space]
      self.board[row][col] = player
      if self.has_player_won(player):
        self.board[row][col] = '-'
        return space
      self.board[row][col] = '-'
    
    # check if next move is winning for opponent
    # check if putting sign in any empty spaces makes has_player_won function True for opponent sign 
    for space in self.empty_spaces:
      (row, col) = self.locations[space]
      self.board[row][col] = 'X'
      if self.has_player_won('X'):   
        self.board[row][col] = '-'
        return space
      self.board[row][col] = '-'
    
    # check if any of empty spaces is in corners set
    # check if any of empty spaces is in centre set
    # check if any of empty spaces is in sides set
    for loc_set in self.loc_sets:
      for space in self.empty_spaces:
        if space in loc_set:
          return space
      
    

  def start(self):
    self.create_board()
    
    # select the first turn of the player randomly
    player = 'X' if self.get_random_first_player() == 1 else '0'

    while 1:
      
      # print which player's turn it is
      print(f"Player {player}'s turn")

      if player == 'X':
        
        # show board to user to select spot of next move
        self.show_board()

        # ask the user to enter row and col number as input
        loc = int(input("Enter square number from 1-9 counting left to right to add your sign: "))
        
        while loc not in self.empty_spaces:
          print(f"Space is filled, choose an empty space")
          loc = int(input("Enter square number from 1-9 counting left to right to add your sign: "))
        print()

      elif player == '0':
        
        # loc = self.computer_random_move() # EASY LEVEL
        loc = self.computer_ai_move(player) # HARD LEVEL
        
        while loc not in self.empty_spaces:
          loc = random.randint(1,9)
        print()
        print('Player 0 has played move')
        print()
        

      # update spot to respective player sign
      self.add_sign(loc, player)

      # check whether currnet player has won game or not - if so print message and break loop
      if self.has_player_won(player):
        print(f"Player {player} has won the game!")
        break

      # check whether board is filled or not - if so print message and break loop
      if self.is_board_full():
        print(f"Draw")
        break

      # swap player's turns if board has not been filled and curr play was not a winning play
      player = self.next_turn(player)
  
    # show final state of board
    print()
    self.show_board()

# start game
tic_tac_toe = TicTacToe()
tic_tac_toe.start()






```
